```python
#remplazar datos nulos por 'Desconocido'
```

```python
for obj in ['MSZoning','MasVnrType','BsmtQual','BsmtCond','BsmtFinType1','BsmtFinType2','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature','BsmtExposure','Exterior1st','Exterior2nd','Electrical','Functional','Utilities','KitchenQual','KitchenQual','SaleType']:
    df[obj].fillna('Desconocido', inplace = True)

df.to_csv('House_Prices.csv', index=True)
df.to_csv('House_Prices_Advanced.csv', index=False)
```

```python
#remplazar datos nulos por 'moda'
```

```python
columnas = ['MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd', 'KitchenQual', 'Functional', 'SaleType']
valor_desconocido = 'Desconocido' 

for columna in columnas:
    moda_columna = df[df[columna] != valor_desconocido][columna].mode()[0]
    df[columna] = df[columna].replace(valor_desconocido, moda_columna)

df.to_csv('House_Prices_Advanced.csv', index=False)
```

```python
#remplazar datos nulos por 'moda'
```

```python
moda_neighborhood = df.loc[df['Neighborhood'] != 'Desconocido', 'Neighborhood'].mode()[0]
df['Neighborhood'] = df['Neighborhood'].replace('Desconocido', moda_neighborhood)

df.to_csv('House_Prices.csv', index=False)
df.to_csv('House_Prices_Advanced.csv', index=False)

#---------------------------------------------------------------------------------------------------
df_advanced = pd.read_csv('House_Prices_Advanced.csv')


```

```python
if 'Alley' in df.columns:
    df = df.drop(columns=['Alley'])

df.to_csv('House_Prices.csv', index=False)
df.to_csv('House_Prices_Advanced.csv', index=False)
```
